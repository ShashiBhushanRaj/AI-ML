```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
sns.set()
```

```python
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.pipeline import make_pipeline
```

```python
def pr(degree=2,**kwargs):
  p=make_pipeline(PolynomialFeatures(degree),LinearRegression(**kwargs))
  return p
```

```python
x=np.arange(100)/100
y=3*(x**3)-10*(x**2)+34
y_pred=pr(2).fit(x[:,np.newaxis],y).predict(x[:,np.newaxis])
```

```python
plt.scatter(x,y)
plt.plot(x,y_pred,color='red')
```

```python
x=np.arange(100)/100
y=3*(x**3)-10*(x**2)+34+(0.5)*np.random.randn(x.size)
y_pred=pr(2).fit(x[:,np.newaxis],y).predict(x[:,np.newaxis])
```

```python
plt.scatter(x,y)
```

```python
from sklearn.model_selection import validation_curve
```

```python
from sklearn.model_selection import validation_curve
import numpy as np

degree = np.arange(0, 15)  # Start from 1, degree=0 means constant model

train_score, val_score = validation_curve(
    pr(), x[:,np.newaxis], y,
    param_name='polynomialfeatures__degree',
    param_range=degree,
    cv=10,
    scoring='r2'
)
```

```python
plt.plot(degree,np.median(train_score,1),color='green',label="TRS")
plt.plot(degree,np.median(val_score,1),color='red',label='VS')
plt.legend(loc='best')
plt.ylim(0,1)
plt.xlabel('degree')
plt.ylabel('score')
```
