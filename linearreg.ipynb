```python
### modeling a bank account

## define a class for bank account
class Bank_acc:
    def __init__(self,owner,balance=0):
        self.owner=owner
        self.balance=balance
    
    def deposit(self,total_amount):
        self.balance+=total_amount
        print(f"{total_amount} is deposited.new balance is {self.balance}")
    
    def withdraw(self,total_amount):
        if total_amount>self.balance:
            print("error,amount is greater than balance")
        else:
            self.balance-=total_amount
            print(f"{total_amount} is withdrawn .new baalnce is {self.balance}")
    def get_balance(self):
        return self.balance
            
acc_t=Bank_acc("shashi",5000)
print(acc_t.balance)
acc_t.deposit(1000)
acc_t.withdraw(600)
print(acc_t.get_balance)
        
```

```python
## practice account
class bankAccount:
    def __init__(self,owner,balance=0):
        self.owner=owner
        self.balance=balance
    
    def deposit(self,amount):
        self.balance+=amount
        print(f"{amount} is deposited .new balance is {self.balance}")
    
    def withdraw(self,amount):
        if amount>self.balance:
            print("error ,amount is greater than the balance")
        else:
            self.balance-=amount
            print(f"{amount} is withdrawn . new balance is {self.balance}")
    def get_balance(self):
        return self.balance
    
    
owner1=bankAccount("shashi",600)
print(owner1.owner)
print(owner1.balance)
owner1.deposit(5000)
print(owner1.balance)
owner1.withdraw(400)
print(owner1.balance)
print(owner1.get_balance)
```
