```python
### Lemmatiaing
from nltk.stem import WordNetLemmatizer
```

```python
import nltk
nltk.download('wordnet')
nltk.download('omw-1.4')   # recommended for full support
```

```python
lemmatizer=WordNetLemmatizer()
```

```python
lemmatizer.lemmatize("going",pos='v')
```

```python
words=['eating','eats','eaten','writting','writes','programming','programs','history','finally','finalize']
```

```python
for word in words:
    print(word+"---->"+lemmatizer.lemmatize(word,pos='v'))
```
