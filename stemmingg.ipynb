```python
words=['eating','eats','eaten','writting','writes','programming','programs','history','finally','finalize']
```

### PorterStemmer

```python
from nltk.stem import PorterStemmer
```

```python
stemming=PorterStemmer()
```

```python
for word in words:
    print(word+"-------->"+stemming.stem(word))
```

### RegexpStemmer Class

```python
from nltk.stem import RegexpStemmer
```

```python
reg_stemmer=RegexpStemmer('ing|s$|e$|able',min=4)
```

```python
reg_stemmer.stem('ableeatable')
```

### Snowball Stemmer

```python
from nltk.stem import SnowballStemmer
```

```python
Snowbal=SnowballStemmer('english')
```

```python
for word in words:
    print(word+"----->"+Snowbal.stem(word))
```

```python
stemming.stem("fairly"),stemming.stem("sportingly")
```

```python
Snowbal.stem("fairly"),Snowbal.stem("sportingly")
```
